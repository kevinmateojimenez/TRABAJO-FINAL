Bajo esta regla de negocio, **el último registro ingresado se considera la versión definitiva y verídica**.

A continuación, aislaremos estos casos para observar las discrepancias (qué datos cambió el estudiante) y procederemos a conservar únicamente la última actualización, descartando los intentos previos.

#### 4.3.2 Imposibilidades Físicas y Solapamiento de Actividades (Multitasking)

Una validación crucial en encuestas de uso del tiempo es verificar los límites físicos inquebrantables del día (24 horas). En este dataset contamos con cuatro variables de tiempo: horas de estudio (`daily_study_hours`), horas de sueño (`daily_sleep_hours`), tiempo en pantallas (`screen_time_hours`) y actividad física (`physical_activity_hours`).

Un análisis superficial sumaría estas cuatro variables de forma lineal. Sin embargo, desde el conocimiento del dominio, sabemos que para un estudiante contemporáneo, **el tiempo de estudio y el tiempo de pantalla pueden ser actividades paralelas** (ej. estudiar programación en un monitor). 

Por lo tanto, para auditar verdaderas imposibilidades físicas, calcularemos el **Límite Físico Estricto**, asumiendo un escenario de solapamiento del 100% entre estudio y pantallas. La fórmula será:
`Límite Estricto = Sueño + Actividad Física + MAX(Estudio, Pantallas)`

Cualquier registro donde este *Límite Estricto* supere las 24 horas representa una imposibilidad biológica y un fallo crítico en la captura del dato.

```python
# 1. Calcular el tiempo máximo concurrente entre uso de pantallas y estudio
tiempo_digital_maximo = df[['daily_study_hours', 'screen_time_hours']].max(axis=1)

# 2. Calcular el Límite Físico Estricto
df['strict_physical_hours'] = df['daily_sleep_hours'] + df['physical_activity_hours'] + tiempo_digital_maximo

# 3. Aislar a los registros que infringen las leyes de la física (los verdaderos imposibles)
improbables_fisicos = df[df['strict_physical_hours'] > 24]

# 4. Visualizar los hallazgos
display(Markdown(f"##### ⏳ Inconsistencia Temporal Crítica Detectada"))
display(Markdown(f"> Incluso asumiendo que los estudiantes estudian y miran pantallas al mismo tiempo, se identificaron **{len(improbables_fisicos)} estudiantes** cuyas rutinas exceden el límite de 24 horas."))

if not improbables_fisicos.empty:
    def resaltar_exceso(val):
        color = '#ffb3b3' if val > 24 else ''
        return f'background-color: {color}; font-weight: bold;'
    
    columnas_mostrar = ['student_id', 'daily_sleep_hours', 'physical_activity_hours', 'daily_study_hours', 'screen_time_hours', 'strict_physical_hours']
    
    # Mostrar los casos más extremos ordenados por el exceso de tiempo
    display(improbables_fisicos.sort_values(by='strict_physical_hours', ascending=False)[columnas_mostrar].head(10).style.map(resaltar_exceso, subset=['strict_physical_hours']))

# ==============================================================================
# ACCIÓN FORENSE: ETIQUETADO DE CALIDAD DE DATOS
# ==============================================================================

# Marcar los datos viables frente a los imposibles
df['time_consistency'] = 'Valid'
df.loc[df['strict_physical_hours'] > 24, 'time_consistency'] = 'Impossible'

display(Markdown(f"**Tratamiento aplicado:** Se ha creado la etiqueta analítica `time_consistency`. Los registros clasificados como 'Impossible' evidencian datos basura o sesgos severos de auto-reporte que no pueden ser explicados por el solapamiento de actividades."))
```

> **Veredicto y Decisión de Tratamiento:**
> 
> Detectar registros que violan las leyes de la física (incluso asumiendo multitasking) plantea un dilema analítico. La acción estándar sería la eliminación (Listwise Deletion). Sin embargo, desde una perspectiva de **psicología de datos**, la sobreestimación del tiempo invertido (percibir que se estudia/trabaja más horas de las reales) es un síntoma clásicamente documentado en individuos con altos niveles de ansiedad o Síndrome de Burnout.
> 
> **Decisión:** 
> > 1. **No se eliminarán** estos registros en la fase actual.
> > 2. Se utilizará la etiqueta creada (`time_consistency`) durante el Análisis Bivariado para validar la *Hipótesis de Distorsión por Estrés* (es decir, comprobar si el grupo `Impossible` presenta una tasa estadísticamente mayor de Burnout).
> > 3. De cara a un futuro modelo predictivo, estos valores de tiempo anómalos serán normalizados (escalados a un máximo de 24h) o tratados mediante imputación, pero la "bandera" de exageración se mantendrá como una variable predictora (Feature Engineering).

#### 4.3.3 Matriz de Coherencia Relacional: Edad vs. Progreso Académico

En la calidad de datos, un valor puede ser estadísticamente normal de forma aislada, pero lógicamente imposible al cruzarlo con otra variable. Al analizar las variables `age` (Edad) y `year` (Año de estudio), debemos aplicar reglas de negocio estrictas basadas en la cronología educativa estándar. 

Adaptando nuestra lógica al contexto demográfico real (donde el ingreso universitario a los 15 años es un escenario factible), se definió la siguiente **Matriz de Coherencia Flexible**:
* `1st` (1er año): Edad mínima = 15 años.
* `2nd` (2do año): Edad mínima = 16 años.
* `3rd` (3er año): Edad mínima = 17 años.
* `4th` (4to año): Edad mínima = 18 años.

Cualquier registro que viole esta escalera cronológica (por ejemplo, un estudiante de 17 años en 4to año) representa una falla inequívoca de integridad relacional (ruido en el auto-reporte o error de sistema) y será intervenido.

```python
# 1. Definir la matriz con parámetros sumamente flexibles
reglas_edad_minima = {
    '1st': 15,
    '2nd': 16,
    '3rd': 17,
    '4th': 18
}

# 2. En lugar de usar máscaras de Pandas directamente, recolectar los IDs o índices anómalos en una lista
indices_anomalos = []

# Iterar de forma segura
for anio, edad_minima in reglas_edad_minima.items():
    # Filtrar a los que rompen la regla
    filtro = (df['year'] == anio) & (df['age'] < edad_minima)
    # Extraer los índices reales y guardarlos en la lista
    indices_anomalos.extend(df[filtro].index.tolist())

# 3. Aislar los registros defectuosos usando loc
df_precoces = df.loc[indices_anomalos]

# 4. Presentar la evidencia
display(Markdown(f"##### 🎓 Auditoría de Escalafón Académico"))
display(Markdown(f"> Se identificaron **{len(df_precoces)} registros** donde la edad reportada es biológicamente inconsistente con el avance académico."))

if not df_precoces.empty:
    display(Markdown("**Muestra de los registros inconsistentes (Evidencia previa a la imputación):**"))
    columnas_ver = ['student_id', 'age', 'course', 'year', 'cgpa']
    display(df_precoces[columnas_ver].head(10).style.map(lambda x: 'background-color: #ffb3b3', subset=['age', 'year']))


if not df_precoces.empty:
    for anio, edad_minima in reglas_edad_minima.items():
        filtro_anomalia = (df['year'] == anio) & (df['age'] < edad_minima)
        
        if filtro_anomalia.sum() > 0:
            # Calcular el valor mínimo condicional
            nuevo_valor = df[(df['year'] == anio) & (df['age'] >= edad_minima)]['age'].min()
            
            # Imputar los valores quirúrgicamente
            df.loc[filtro_anomalia, 'age'] = nuevo_valor
```

> **Tratamiento aplicado:** Para mantener la política de "Cero Nulos" sin alterar la distribución general, se aplicó un algoritmo de **Imputación Condicional Dinámica**. Los infractores de cada año recibieron la edad mínima exclusiva de sus compañeros de corte que sí cumplían con la regla cronológica. Esto neutraliza el ruido sin castigar la retención de datos vitales para el análisis psicológico.

### 4.3.4 Ruido de Rango y Outliers Extremos: Anomalías Demográficas (Edad)

Una vez resueltas las inconsistencias lógicas cruzadas y las imposibilidades físicas de tiempo, nos enfocamos en el ruido de rango de variables individuales. En este caso, la variable `age` (edad).

En el contexto universitario general, las edades tienen una distribución asimétrica positiva muy marcada: la inmensa mayoría se concentra entre los 18 y los 25 años, con una cola larga hacia la derecha (estudiantes de posgrado o adultos retomando estudios). Sin embargo, la presencia de edades correspondientes a la tercera edad (ej. > 60, 80 o 90 años) rompe el sentido común demográfico para este conjunto de datos. 

Estos valores no representan la variabilidad natural de la población, sino "ruido" puro (errores tipográficos durante la captura de datos, como digitar "85" en lugar de "18" o "25"). Dejarlos intactos deformaría la media, arruinaría las visualizaciones univariadas y sesgaría cualquier modelo de machine learning posterior.

A continuación, exponemos la gravedad de estos valores atípicos.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Resumen estadístico de la edad para ver el máximo disparatado
print("Resumen estadístico de la variable 'age':")
display(df[['age']].describe().T)

# 2. Visualización del problema: Boxplot + Histograma
fig, (ax_box, ax_hist) = plt.subplots(2, 1, sharex=True, figsize=(12, 6), gridspec_kw={"height_ratios": (.15, .85)})
sns.set_style("whitegrid")

# Boxplot arriba (para detectar los puntos atípicos)
sns.boxplot(x=df['age'], ax=ax_box, color='#FC9C28') # Naranja UAL
ax_box.set(xlabel='')
ax_box.set_title('Distribución de la Edad y Outliers Extremos', fontsize=14)

# Histograma abajo (para ver la concentración real vs el ruido)
sns.histplot(df['age'], bins=30, kde=False, ax=ax_hist, color='#BA352C') # Rojo UAL
ax_hist.set(xlabel='Edad (Años)', ylabel='Cantidad de Estudiantes')

plt.tight_layout()
plt.show()

# 3. Filtrar y exhibir a los "ancianos universitarios"
umbral_edad_extrema = 70 # Ajustable según tu criterio
df_edades_absurdas = df[df['age'] > umbral_edad_extrema]

print(f"\nTotal de estudiantes con edades irreales (> {umbral_edad_extrema} años): {len(df_edades_absurdas)}")
display(df_edades_absurdas[['student_id', 'age', 'course', 'year', 'burnout_level']].head(10))
```

> **Tratamiento aplicado:** Para aislar las anomalías demográficas sin recurrir a la eliminación destructiva, se implementó una estrategia de etiquetado (*Flagging*) basada en un método paramétrico estándar. Se definió el límite de normalidad superior utilizando la fórmula de la media poblacional más $5$ veces la desviación estándar ($\mu + 5\sigma$). Todo registro que superó este umbral recibió la etiqueta `'Suspicious'` en una nueva variable de control (`is_age_suspicious`).
>
> **Justificación del Negocio:** La edad universitaria suele tener una distribución asimétrica con una cola larga hacia la derecha. Al fijar el límite en 5 desviaciones estándar por encima de la media, garantizamos la protección de los estudiantes "fuera de la edad tradicional" (adultos, posgrados o reingresos), asegurando que la bandera de error señale única y exclusivamente a valores demográficos físicamente inverosímiles (errores de captura en el sistema).

```python
# 1. Calcular la Media poblacional y la Desviación Estándar para 'age'
edad_media = df['age'].mean()
edad_std = df['age'].std()

# 2. Calcular el umbral estadístico superior (Media + 5 Desviaciones Estándar)
# Mantenemos el multiplicador en 5 para ser extremadamente conservadores
umbral_superior = edad_media + (5 * edad_std)

# 3. Etiquetar los registros que superen el umbral en una nueva columna de control (flag)
df['is_age_suspicious'] = np.where(df['age'] > umbral_superior, 'Suspicious', 'Normal')

# 4. Verificar el resultado
print(f"El umbral superior estadístico calculado es: {umbral_superior:.2f} años\n")

# Mostrar una muestra de los infractores
df_sospechosos = df[df['is_age_suspicious'] == 'Suspicious']
print(f"Total de registros marcados como sospechosos: {len(df_sospechosos)}")
display(df_sospechosos[['student_id', 'age', 'course', 'year', 'is_age_suspicious']].head(10))
```

### 4.3.5 Inconsistencias Categóricas (Falta de Estandarización)

Las variables cualitativas (texto) son altamente susceptibles a errores de formato, especialmente si provienen de captura manual o de la consolidación de múltiples sistemas. Las variaciones de mayúsculas/minúsculas y los espacios en blanco invisibles (trailing/leading spaces) generan una fragmentación artificial de las categorías.

Si no se estandariza el texto, el motor analítico tratará variaciones semánticamente idénticas (ej. `'Male'` y `'Male '`) como entidades distintas. Esto arruinaría la integridad de los conteos y la posterior visualización univariada.

A continuación, ejecutamos una auditoría de distribución de frecuencias sobre todas las variables categóricas. Esto nos permite identificar no solo la presencia de ruido tipográfico, sino también el peso porcentual de cada inconsistencia dentro del conjunto de datos.

```python

# 1. Identificar automáticamente las columnas categóricas
columnas_categoricas = df.select_dtypes(include=['object', 'str']).columns

print("--- AUDITORÍA DE CATEGORÍAS (VISTA PIVOTEADA) ---\n")

for col in columnas_categoricas:
    # 2. Calcular Frecuencias y Porcentajes
    frec = df[col].value_counts(dropna=False)
    porc = df[col].value_counts(dropna=False, normalize=True) * 100
    
    # 3. Visibilizar espacios en blanco envolviendo los nombres en comillas
    nombres_columnas = [f"'{x}'" if pd.notnull(x) else 'NaN' for x in frec.index]
    
    # 4. Construir la tabla pivoteada (2 filas, N columnas)
    tabla = pd.DataFrame(
        [frec.values, porc.values],
        index=['Frecuencia', 'Porcentaje'],
        columns=nombres_columnas
    )
    
    # 5. Aplicar estilos avanzados por filas usando pd.IndexSlice
    estilo_tabla = (tabla.style
        # Formato de números: enteros para la fila 1, porcentajes para la fila 2
        .format("{:.0f}", subset=pd.IndexSlice['Frecuencia', :])
        .format("{:.2f}%", subset=pd.IndexSlice['Porcentaje', :])
        # Aplicar la barra SOLO a la fila de porcentaje. axis=1 le dice a Pandas que lo haga horizontalmente
        .bar(subset=pd.IndexSlice['Porcentaje', :], axis=1, color='rgba(252, 156, 40, 0.3)', vmin=0, vmax=100)
        # Título con el color exacto solicitado
        .set_caption(f"Distribución en: {col}")
        .set_table_styles([{
            'selector': 'caption',
            'props': [('font-size', '15px'), ('color', '#BA352C'), ('font-weight', 'bold'), ('text-align', 'left'), ('margin-bottom', '8px')]
        },
        {   # Para darle un poco de estilo al nombre de las filas (Frecuencia / Porcentaje)
            'selector': 'th.row_heading',
            'props': [('font-weight', 'bold')]
        }])
    )
    
    # 6. Desplegar la tabla
    display(estilo_tabla)
    print("\n")
```

> **Tratamiento a aplicar:** Se implementó una rutina de limpieza vectorizada y condicional sobre todas las variables categóricas. El proceso constó de tres pasos en un solo ciclo: (1) Eliminación de espacios residuales (`strip`) en todas las columnas. (2) Para la variable `course`, se aplicó una homogenización a mayúsculas seguida de un mapeo por diccionario para restaurar y proteger la nomenclatura académica estandarizada (ej. `BTech`, `BSc`). (3) Para el resto de variables cualitativas, se forzó un formato de título (`Title Case`) para colapsar variaciones de captura (ej. `"MALE"`, `"male"` $\rightarrow$ `"Male"`).
>
> **Justificación del Negocio:** La consolidación de categorías es el prerrequisito innegociable del análisis univariado. Si no se colapsan los errores tipográficos, el motor analítico fragmentará artificialmente a la población, arruinando los conteos de frecuencia y diluyendo la representatividad estadística de variables críticas como los niveles de burnout, estrés o calidad de sueño.

```python
# 1. Identificar todas las columnas categóricas (compatibilidad Pandas 2/3)
columnas_categoricas = df.select_dtypes(include=['object', 'str']).columns

# 2. Diccionario de mapeo estricto para las nomenclaturas académicas de la variable 'course'
# (Aseguramos las carreras principales que vimos en la auditoría)
mapeo_carreras = {
    'BTECH': 'BTech',
    'B-TECH': 'BTech',
    'MTECH': 'MTech',
    'BSC': 'BSc',
    'MSC': 'MSc',
    'BCA': 'BCA',
    'B.C.A.': 'BCA',
    'MCA': 'MCA',
    'BBA': 'BBA',
    'M.B.A': 'MBA',
    'OTHER': 'Other'
}

# 3. Aplicar limpieza vectorizada en un solo paso
for col in columnas_categoricas:
    # A. Aniquilar espacios invisibles al inicio y final en absolutamente todas las columnas
    df[col] = df[col].str.strip()
    
    # B. Tratamiento condicional según el dominio de la variable
    if col == 'course':
        # Estandarizar temporalmente todo a mayúsculas para evitar fugas por minúsculas rebeldes
        df[col] = df[col].str.upper()
        # Aplicar el reemplazo quirúrgico para recuperar la estética académica
        df[col] = df[col].replace(mapeo_carreras)
    else:
        # Formato de título estándar para variables regulares (Gender, Stress Level, etc.)
        df[col] = df[col].str.title()

# 4. Verificación final de integridad (Auditoría post-limpieza)
print("--- RESULTADO DE LA ESTANDARIZACIÓN GLOBAL ---\n")
for col in columnas_categoricas:
    valores_limpios = df[col].dropna().unique()
    print(f"✅ '{col}' ({len(valores_limpios)} categorías): {list(valores_limpios)}")
```
