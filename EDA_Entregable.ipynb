# Análisis Exploratorio de Datos (EDA)
## Salud mental estudiantil y síndrome de burnout

### 📄Introducción

<p>
    El Conjunto de Datos sobre Salud Mental y Agotamiento Estudiantil (150k filas, 20 características) es una base de datos a gran escala diseñada para analizar y predecir los niveles de burnout (agotamiento) estudiantil basándose en factores académicos, psicológicos y de estilo de vida. 
</p>

    Este conjunto de datos contiene 150,000 registros de estudiantes con una combinación de variables numéricas y categóricas, lo que lo hace ideal para tareas de aprendizaje automático (machine learning), clasificación y análisis de datos.


### 🧮Conjunto de Datos
**Caracteristicas:**
- Demografía: Edad, género, carrera y año académico.

- Indicadores académicos: Promedio general (CGPA), asistencia y horas de estudio.

- Puntajes psicológicos: Niveles de ansiedad, depresión y estrés.

- Factores de estilo de vida: Horas de sueño, actividad física y tiempo frente a la pantalla.

- Indicadores de estrés social y financiero.

- Variable objetivo: `burnout_level` (nivel de agotamiento).

#### Importacion del conjunto de datos

```python
from IPython.core.display import Markdown
from IPython.core.display_functions import display
import pandas as pd

# Cargar dataset
path = "data/datos.csv"
df_raw = pd.read_csv(path, sep=",", header="infer")

# Crear copia de trabajo
df = df_raw.copy()

# Total de filas
total_filas = len(df)
display(Markdown(f"Total de registros: {total_filas}"))
display(Markdown(f"#### **Muestra del conjunto de datos**"))
display(df.sample(n=10, ignore_index=True))
```
