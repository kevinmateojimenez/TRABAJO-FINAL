## 1. 📄 Introducción 

<p>
    El Conjunto de Datos sobre Salud Mental y Agotamiento Estudiantil (150k filas, 20 características) es una base de datos a gran escala diseñada para analizar y predecir los niveles de burnout (agotamiento) estudiantil basándose en factores académicos, psicológicos y de estilo de vida. 
</p>

> Este conjunto de datos contiene 150,000 registros de estudiantes con una combinación de variables numéricas y categóricas, lo que lo hace ideal para tareas de aprendizaje automático (machine learning), clasificación y análisis de datos.

## 2. 🧮 Conjunto de Datos
**Características:**
- Demografía: Edad, género, carrera y año académico.

- Indicadores académicos: Promedio general (CGPA), asistencia y horas de estudio.

- Puntajes psicológicos: Niveles de ansiedad, depresión y estrés.

- Factores de estilo de vida: Horas de sueño, actividad física y tiempo frente a la pantalla.

- Indicadores de estrés social y financiero.

- Variable objetivo: `burnout_level` (nivel de agotamiento).

## 4. ✅ Calidad de los datos

Antes de realizar un análisis exhaustivo de los datos, primero evaluaremos qué tan fiables son, revisando datos vacíos, valores improbables y valores duplicados.

### 4.1 Datos Vacíos

```python
# 1. Calculamos la suma de nulos
nulos = df.isnull().sum()

# 2. Filtramos solo los que son mayores a cero
nulos_filtrados = nulos[nulos > 0]

# 3. Validamos si hay resultados para mostrar
if not nulos_filtrados.empty:
    # Convertimos a DataFrame y luego a markdown
    tabla_nulos_md = nulos_filtrados.to_frame(name="Valores Nulos").to_markdown()
    
    display(Markdown("### Columnas con Valores Nulos"))
    display(Markdown(tabla_nulos_md))
else:
    display(Markdown("### No hay valores nulos en el DataFrame 🎉"))
```

```python
# Filtramos el DataFrame para obtener solo las filas con algún valor nulo
df_filas_nulas = df[df.isnull().any(axis=1)]
# Mostramos el mensaje superior
display(Markdown(f"**Mostrando {len(df_filas_nulas)} filas que contienen al menos un valor nulo:**"))
# Aplicamos el estilo para resaltar los nulos en amarillo y lo mostramos
display(df_filas_nulas.style.highlight_null(color='yellow'))
```
